# Two point model - isothermal case

This notebook compares the diffusive and convective isothermal cases

In both cases, we normalize the density and radial coordinate to the upstream exponential profile scale length $\lambda_u$, so $n_u = exp(-r)$. This gives a parameter $\lambda$ 

|Case|Parameter|
|-|-|
|Diffusive|$\sqrt{D L_\parallel / c_s}/\lambda_u$|
|Convective|$L_\parallel v / c_s/\lambda_u$|

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def nt_diffusive(r,lam):
    return ((2*lam**2-3)*np.exp(-np.sqrt(2)*lam*r) + 2*np.exp(-r))/(4*lam**2-2)

def nt_convective(r,lam):
    return ((2*lam-3)*np.exp(-2*lam*r) + 2*np.exp(-r))/(4*lam-2)
```

```python
def plot_nt(lam=1., r = np.linspace(0,5,1000)):
    plt.figure()
    plt.title(f'diff vs conv lam={lam:.2f}')
    plt.semilogy(r, nt_convective(r,lam),label='conv')
    plt.semilogy(r, nt_diffusive(r,lam),label='diff')
    plt.semilogy(r, np.exp(-r), 'k:', label='n_u')
    plt.semilogy(r, np.exp(-np.sqrt(2)*lam*r)/2, c='gray', ls='--', label='e^(-lambda r)')
    plt.xlabel('r')
    plt.ylabel('n_t')
    plt.legend()
    plt.ylim([1e-4,2])
    
L = [0.1, 1., np.sqrt(3/2), 3/2, 5]
for lam in L:
    plot_nt(lam=lam)
```
