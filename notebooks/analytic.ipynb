# Analytic computations for the two-point model

```python
import sympy as s
```

## Isothermal, diffusive

```python
r,λ = s.symbols('r λ',real=True, positive=True)
n_td = s.Function('n_td')
eqd = s.Derivative(n_td(r),(r,2))-2*λ**2*n_td(r)+s.exp(-r)
s.dsolve(eqd,n_td(r),ics={n_td(0):s.Rational(1/2), n_td(s.oo): 0})
```

## Isothermal, convective

```python
n_tc = s.Function('n_tc')
eqc = s.Derivative(n_tc(r),r)+2*λ*n_tc(r)-s.exp(-r)
s.dsolve(eqc,n_tc(r),ics={n_tc(0):s.Rational(1/2)})
```
